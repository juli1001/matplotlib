```python
# another data base

heart_disease=pd.read_csv("/content/heart-disease.csv")
heart_disease
```

```python
# subplots
heart_disease.plot.hist(figsize=(10,10),subplots=True);
```

```python
# plotting with pandas using the oo method

over_50=heart_disease[heart_disease["age"]>50]
```

```python
over_50
```

```python
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax)
ax.set_xlim([45,100]);
```

```python
# more complex graph
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(over_50["age"],over_50["chol"],c=over_50['target']);

# customize graph

ax.set(title='Heart Disease and Cholesterol levels',xlabel="Age",ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(),title="Target")

plt.figure(figsize=(10,5))
plt.scatter(over_50.age[over_50.target==1],
            over_50.chol[over_50.target==1],c="red")
plt.scatter(over_50.age[over_50.target==0],
            over_50.chol[over_50.target==0],c='green')

plt.title('Heart Disease and Cholesterol levels')
plt.xlabel('Age')
plt.ylabel("Chol")
plt.legend(["Disease","No Disease"]);
```

```python
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(over_50["age"],over_50["chol"],c=over_50['target']);

# customize graph

ax.set(title='Heart Disease and Cholesterol levels',xlabel="Age",ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(),title="Target")

#Add a meanline
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
# adding another plot to existing styled one
fig,(ax0,ax1)=plt.subplots(ncols=1,nrows=2,sharex=True,figsize=(10,6))

scatter=ax0.scatter(over_50["age"],over_50['chol'],c=over_50['target'])

ax0.set(title="Heart Disease and Cholesterol Levels", ylabel="Cholesterol");
ax0.legend(*scatter.legend_elements(),title="Target");
ax0.axhline(y=over_50["chol"].mean(),c="b",linestyle="--",label="Average")

scatter=ax1.scatter(over_50["age"],over_50['thalach'],c=over_50['target'])

ax1.set(title="Heart Disease and Max Heart Rate Levels", ylabel="Max Heart Rate");
ax1.legend(*scatter.legend_elements(),title="Target");
ax1.axhline(y=over_50["thalach"].mean(),c="b",linestyle="--",label="Average")

# title figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold")
```
